```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json

# Create lists to store the data
Noms = []
Adresses = []
Numeros = []
Secteurs = []

# List of countries with their codes
pays = ['bj', 'bf', 'ci', 'ml', 'ne', 'sn', 'tg', 'cm', 'cf', 'cg', 'dj', 'ga', 'mg', 'rw', 'gn', 'mr', 'cd', 'td']

secteurs = [
    'Administrations',
    'Communication',
    'Comptabilité, conseil & juridique',
    'Bâtiment & Construction',
    'Automobile',
    'Commerces',
    'Informatique',
    'Formation, éducation',
    'Santé',
    'Télécommunication',
    'Culture, tourisme & Loisirs',
    'Transport & Voyage',
    'Services',
    'Sport',
    'Autres',
    'Immobilier',
    'Agroalimentaire',
    'Finances',
    'Industrie',
    'Énergie',
    'Sécurité'
]

import re

for code in pays:
    for secteur in secteurs:
        # Adjust the number of pages as needed
        url = f"https://showroomafrica.com/{code}/rechercher-entreprise?nom=&secteur={secteur}"
        response = requests.get(url)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            # Retrieve all occurrences of the <div class="company-info">
            company_info_list = soup.find_all('div', class_='company-info')

            for company_info in company_info_list:
                # Retrieve the information from each occurrence
                company_name_element = company_info.select_one('.company-name a')
                company_name = company_name_element.text.strip() if company_name_element else None

                company_category_element = company_info.select_one('.company-category')
                company_category = company_category_element.text.strip() if company_category_element else None

                address_element = company_info.select_one('.contacts li:nth-of-type(1)')
                address = address_element.text.strip() if address_element else None

                # Check if phone number exists before extracting
                phone_number_element = company_info.select_one('.contacts li:nth-of-type(2) b')
                phone_number = phone_number_element.text.strip() if phone_number_element else None

                # Append the extracted information to the lists
                Noms.append(company_name)
                Secteurs.append(company_category)
                Adresses.append(address)
                Numeros.append(phone_number)
```

```python
# Create a DataFrame from the lists
df = pd.DataFrame({
    'Nom': Noms,
    'Secteur': Secteurs,
    'Adresse': Adresses,
    'Numero': Numeros
})

# Save the DataFrame to a CSV file
df.to_csv('entreprise_data.csv', index=False)
```
