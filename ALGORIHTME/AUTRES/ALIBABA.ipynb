```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Créez un DataFrame pour stocker les données
columns = ["Titre", "Image", "Prix", "Fournisseur"]
df = pd.DataFrame(columns=columns)

# Définissez le nombre total de pages que vous souhaitez parcourir
total_pages = 10  # Remplacez 5 par le nombre total de pages que vous souhaitez parcourir

for page in range(1, total_pages + 1):
    url = f"https://www.alibaba.com/premium/ali_baba_s/{page}.html"
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        product_divs = soup.find_all('div', class_='component-offer-product-gallery')

        for product_div in product_divs:
            title = product_div.select_one('.product-title-text').text.strip()
            image_link = product_div.select_one('.product-img')['src']
            price_range = product_div.select_one('.price-number').text.strip()
            price_unit = product_div.select_one('.price-unit').text.strip()
            supplier_name = product_div.select_one('.product-supplier-name').text.strip()

            # Ajoutez une nouvelle ligne au DataFrame
            df = df.append({"Titre": title, "Image": image_link, "Prix": f"{price_range} {price_unit}", "Fournisseur": supplier_name},
                           ignore_index=True)

# Affichez le DataFrame à la fin
print(df)
```
